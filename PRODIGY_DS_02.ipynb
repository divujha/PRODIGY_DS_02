```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("/content/sample_data/train.csv")
```

```python
print(data.head())
```

```python
data.describe()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.dropna(subset=["Embarked"], inplace=True)
data["Cabin"].fillna("Unknown", inplace=True)
data["Age"].fillna(data["Age"].mean(), inplace=True)
```

```python
data.isnull().sum()
```

```python
plt.figure(figsize=(6, 3))
sns.histplot(data["Age"], kde=True)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
sns.barplot(x='Sex', y='Survived', data=data, ci=None)
plt.title('Survival Rate by Gender')
plt.show()
```

```python
sns.barplot(x='Pclass', y='Survived', data=data, ci=None)
plt.title('Survival Rate by Passenger Class')
plt.show()
```

```python
plt.hist(data['Age'], bins=20)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution of Passengers')
plt.show()
```

```python
# Create age groups
data['AgeGroup'] = pd.cut(data['Age'], bins=[0, 18, 30, 50, np.inf], labels=['Child', 'Young Adult', 'Adult', 'Senior'])

sns.barplot(x='AgeGroup', y='Survived', data=data, ci=None)
plt.title('Survival Rate by Age Group')
plt.show()
```

```python
plt.figure(figsize=(6, 3))
sns.scatterplot(data=data, x="Age", y="Fare", hue="Survived")
plt.title("Scatter Plot of Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.legend(title="Survived")
plt.show()
```
